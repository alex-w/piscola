## DES fitted samples (AllBands, Opt[g, r, i, z] & IR[z, Y, J, H, K])

We are going to obtain a list of all the SNe with a "successful" fit for AllBands, Opt & IR.

```python
source = '/home/tem1u17/Desktop/Research/DES/SNoopy_Fits/'
sn_list_allbands = [d.split('_')[0] for d in os.listdir(source) 
                    if (os.path.isfile(os.path.join(source, d)) and d[-8:]=='plot.png')]
sn_list_opt = [d.split('_')[0] for d in os.listdir(source) 
                    if (os.path.isfile(os.path.join(source, d)) and d[-12:]=='plot_opt.png')]
sn_list_ir = [d.split('_')[0] for d in os.listdir(source) 
                    if (os.path.isfile(os.path.join(source, d)) and d[-11:]=='plot_ir.png')]

print("The number of successful fits were {:.0f} SNe for AllBands, {:.0f} SNe for Opt and {:.0f} SNe for IR".
      format(len(sn_list_allbands), len(sn_list_opt), len(sn_list_ir)))
```

We will compare which SNe are in which list.

```python
sn_list_allbands_not_ir = [sn for sn in sn_list_allbands if sn not in sn_list_ir]
sn_list_allbands_not_opt = [sn for sn in sn_list_allbands if sn not in sn_list_opt]
sn_list_opt_not_ir = [sn for sn in sn_list_opt if sn not in sn_list_ir]
sn_list_ir_not_opt = [sn for sn in sn_list_ir if sn not in sn_list_opt]
sn_list_ir_not_allbands = [sn for sn in sn_list_ir if sn not in sn_list_allbands]
sn_list_opt_not_allbands = [sn for sn in sn_list_opt if sn not in sn_list_allbands]
print(sn_list_allbands_not_ir)
print(sn_list_allbands_not_opt)
print(sn_list_opt_not_ir)
print(sn_list_ir_not_opt)
print(sn_list_ir_not_allbands)
print(sn_list_opt_not_allbands)
```

Let's find out which bands are the ones with fits

```python
import mmap

f = open('sn_list_ir_not_allbands.txt','w')
bands = ['des_g', 'des_r', 'des_i', 'des_z', 'vista_Y', 'vista_J', 'vista_H', 'vista_K']
for sn in sn_list_ir_not_allbands:
    file = source+sn+'.txt'
    band_list = [sn]
    with open(file, 'rb', 0) as file, mmap.mmap(file.fileno(), 0, access=mmap.ACCESS_READ) as s:
        for band in bands:
            band_bits = band.encode(encoding='utf-8') #mmap works with bytes-like objects instead of strings
            if s.find(band_bits) != -1:
                band_list.append(band)
            else:
                band_list.append('-------')
    print(band_list)
    for item in band_list:
        f.write("%s\t" % item)
    f.write('\n')
f.close()    
```

## Model Residuals

```python
source_des = '/home/tem1u17/Desktop/Research/DES/SNoopy_Fits/'
test_params = pd.read_csv(source_des + 'SNe_test2.dat', sep=' ')
test_params.head(n=5)
```

```python
bands = ['_g', '_r', '_i', '_z', '_Y', '_J', '_H', '_K']
bands= ['_g']

f, ax = plt.subplots(figsize=(6, 6))

for sn in test_params['SN'].unique():
#for sn in ['DES15C3axd']:
    sn_params = test_params[test_params['SN'] == sn]
    Tmax = sn_params['Tmax'].unique()
    
    data_mag_cols = [col for col in sn_params.columns if ('mag_des' in col or 'mag_vista' in col) and 'e_mag' not in col]
    data_e_mag_cols = [col for col in sn_params.columns if 'e_mag' in col]
    data_MJD_cols = [col for col in sn_params.columns if 'MJD_des' in col or 'MJD_vista' in col]
    model_mag_cols = [col for col in sn_params.columns if 'model_mag' in col]
    model_MJD_cols = [col for col in sn_params.columns if 'model_MJD' in col]
    
    data_mag = sn_params[data_mag_cols]
    data_e_mag = sn_params[data_e_mag_cols]
    data_MJD = sn_params[data_MJD_cols]
    model_mag = sn_params[model_mag_cols]
    model_MJD = sn_params[model_MJD_cols]
    
    for band in bands:
        d = data_mag[[col for col in data_mag.columns if band in col][0]]
        m = model_mag[[col for col in model_mag.columns if band in col][0]]
        e = data_e_mag[[col for col in data_e_mag.columns if band in col][0]]
        t = data_MJD[[col for col in data_MJD.columns if band in col][0]]
        y = (d-m)/e
        x = t - Tmax
        ax.plot(x, y, 'o')
ax.axhline(y=0.0, color='k', linestyle='-')
plt.title(band[-1] + '-band Residuals', fontsize = 18)
plt.ylabel('(data - model)/error', fontsize = 18)
plt.xlabel('MJD - Tmax(B)', fontsize = 18)
#plt.savefig(band[-1] + '_band _Residuals.png')
```

```python
bands = ['_g', '_r', '_i', '_z', '_Y', '_J', '_H', '_K']


h0 = []
h1 = []
h2 = []
h3 = []
h4 = []
h5 = []
h6 = []
h7 = []

H = [h0, h1, h2, h3, h4, h5, h6, h7]

cuts = np.arange(0 , 80 ,10)
for sn in test_params['SN'].unique():
#for sn in ['DES15C3axd']:
    sn_params = test_params[test_params['SN'] == sn]
    Tmax = sn_params['Tmax'].unique()
    
    data_mag_cols = [col for col in sn_params.columns if ('mag_des' in col or 'mag_vista' in col) and 'e_mag' not in col]
    data_e_mag_cols = [col for col in sn_params.columns if 'e_mag' in col]
    data_MJD_cols = [col for col in sn_params.columns if 'MJD_des' in col or 'MJD_vista' in col]
    model_mag_cols = [col for col in sn_params.columns if 'model_mag' in col]
    model_MJD_cols = [col for col in sn_params.columns if 'model_MJD' in col]
    
    data_mag = sn_params[data_mag_cols]
    data_e_mag = sn_params[data_e_mag_cols]
    data_MJD = sn_params[data_MJD_cols]
    model_mag = sn_params[model_mag_cols]
    model_MJD = sn_params[model_MJD_cols]
    
    for band in bands:
        d = data_mag[[col for col in data_mag.columns if band in col][0]]
        m = model_mag[[col for col in model_mag.columns if band in col][0]]
        e = data_e_mag[[col for col in data_e_mag.columns if band in col][0]]
        t = data_MJD[[col for col in data_MJD.columns if band in col][0]]
        y = ((d-m)/e).dropna()
        x = (t - Tmax).dropna()
        for i, time in enumerate(x):
            if len(y) != 0 and time >= cuts[0]-10 and time < cuts[-1]:
                h_idx = np.where(np.logical_and(cuts-10. <= time, time < cuts))[0][0]
                H[h_idx].append(y.values[i])       
        
```

```python
from math import sqrt
f, ax = plt.subplots(figsize=(10, 10))
labels = []
for i, val in enumerate(cuts):
    h = np.asarray(H[i])
    N = len(h)
    avg = np.around(h.mean(), decimals = 2)
    std_N = np.around(h.std()/sqrt(N), decimals = 2)   
    labels.append('time:[' + str(val-10) +',' + str(val) +'], N='+ str(N) +', avg=' + str(avg)+ ', std/sqrt(N)=' +str(std_N))
    
for i, h in enumerate(H):
    ax.hist(h, density=True, label=labels[i])

ax.axvline(x=5, color='k', linestyle='--')
ax.axvline(x=-5, color='k', linestyle='--')
#ax.axvline(x=10, color='k', linestyle='--')
#ax.axvline(x=-10, color='k', linestyle='--')
plt.xlabel('(data - model)/error', fontsize = 16)
plt.ylabel('density of SNe', fontsize = 16)
ax.legend(bbox_to_anchor=(0.8, 1), loc=2, borderaxespad=0., fontsize = 16)
plt.xlim(-20, 20)
plt.savefig('Residuals_histogram.png')
```

```python
labels = []
for i, val in enumerate(cuts):
    h = np.asarray(H[i])
    avg = np.around(h.mean(), decimals = 2)
    std = np.around(h.std(), decimals = 2)
    labels.append('time:[' + str(val-10) +',' + str(val) +'], N='+ str(len(h)) +'; avg,std=' + str(avg)+ ', ' +str(std))
    
for i, h in enumerate(H):
    f, ax = plt.subplots(figsize=(10, 10))
    ax.hist(h, density=True, label=labels[i])

    ax.axvline(x=5, color='k', linestyle='--')
    ax.axvline(x=-5, color='k', linestyle='--')
    ax.axvline(x=10, color='k', linestyle='--')
    ax.axvline(x=-10, color='k', linestyle='--')
    plt.xlabel('(data - model)/error', fontsize = 16)
    plt.ylabel('density of SNe', fontsize = 16)
    ax.legend()    
```

### Counts of SNe per redshift bin

```python
z_des = DES_params['z']
z1 = bands_params['[z1']
z2 = bands_params['z2)']
idx = np.zeros_like(z1).astype(int)

for z in z_des:
    z_idx = np.where(np.logical_and(z1 <= z, z < z2))[0][0]
    idx[z_idx] += 1
        
bands_params['num_sn'] = idx
```

### Counts of data-points per redshift bin

```python
source_des = '/home/tem1u17/Desktop/Research/DES/SNoopy_Fits/'
test_params = pd.read_csv(source_des + 'SNe_test2.dat', sep=' ')
test_params.head()
```

```python
bands = ['_g', '_r', '_i', '_z', '_Y', '_J', '_H', '_K']
columns = ['g_counts', 'r_counts', 'i_counts', 'z_counts', 'Y_counts', 'J_counts', 'H_counts', 'K_counts']
for col in columns:
    bands_params[col] = 0
    
total_counts = np.zeros([len(bands_params['des_g']), len(bands)]).astype(int)

for sn in test_params['SN'].unique():
#for sn in ['DES15C3axd']:
    bands_counts = []
    sn_params = test_params[test_params['SN'] == sn]
    z = DES_params[DES_params['SN'] == sn ]['z'].values[0]
    data_mag_cols = [col for col in sn_params.columns if ('mag_des' in col or 'mag_vista' in col) and 'e_mag' not in col]
    data_mag = sn_params[data_mag_cols]
    for band in bands:
        counts = len(data_mag[[col for col in data_mag.columns if band in col][0]].dropna().values)
        bands_counts.append(counts)        
    z_idx = np.where(np.logical_and(z1 <= z, z < z2))[0][0]
    total_counts[z_idx] += np.asarray(bands_counts)
for row, values in enumerate(total_counts):
    bands_params.loc[row, columns] = values    
```

## Biases between the "Optic" and the "IR"

```python
source_des = '/home/tem1u17/Desktop/Research/DES/'

DES_params_1 = pd.read_csv(source_des + 'SNoopy_Fits/DES_parameters_OptBands.dat', sep='\t')
data_des_1 = DES_params_1#.drop(['Tmax', 'Ymax', 'Hmax', 'Kmax', 'Jmax', 'umax'], axis=1)
data_des_1 = data_des_1.replace(99.9, np.nan)
data_des_1 =  data_des_1[data_des_1['rchisq'] < 5.0]
data_des_1 = data_des_1.add_suffix('_Opt')
data_des_1 = data_des_1.rename(index=str, columns={"SN_Opt": "SN"})

DES_params_2 = pd.read_csv(source_des + 'SNoopy_Fits/DES_parameters_IRBands.dat', sep='\t')
data_des_2 = DES_params_2#.drop(['Tmax', 'Ymax', 'Hmax', 'Kmax', 'Jmax', 'umax'], axis=1)
data_des_2 = data_des_2.replace(99.9, np.nan)
data_des_2 =  data_des_2[data_des_2['rchisq'] < 5.0]
data_des_2 = data_des_2.add_suffix('_IR')
data_des_2 = data_des_2.rename(index=str, columns={"SN_IR": "SN"})
```

```python
result = pd.merge(data_des_1, data_des_2, on='SN')
result
```

```python
DES_params_3 = pd.read_csv(source_des + 'SNoopy_Fits/DES_parameters_AllBands.dat', sep='\t')
data_des_3 = DES_params_3#.drop(['Tmax', 'Ymax', 'Hmax', 'Kmax', 'Jmax', 'umax'], axis=1)
data_des_3 = data_des_3.replace(99.9, np.nan)
data_des_3 =  data_des_3[data_des_3['rchisq'] < 5.0]
data_des_3 = data_des_3.add_suffix('_All')
data_des_3 = data_des_3.rename(index=str, columns={"SN_All": "SN"})

result2 = pd.merge(result, data_des_3, on='SN')
result2
```

```python
param = 'gmax'
x_label, y_label, z_label = param + '_All', param + '_Opt', param + '_IR'
x, y, z = result2[x_label], result2[y_label], result2[z_label]

f, ax = plt.subplots(figsize=(6, 6))
ax.plot(x, z, 'o', label = param)
ax.plot(ax.get_xlim(), ax.get_ylim(), ls="--", c=".3")
ax.legend()
print(len(x.dropna()), len(y.dropna()), len(z.dropna()))
```
